## class 기본 사용법

```python
# 기대 수명 속성을 추가합니다.
#  info라는 메서드를 추가해서, 동물의 이름과 기대 수명을 출력합니다.
# 동물 게임할 때, 요 info 메서드 호출해서 이름이랑 기대수명 출력
# from typing import List, Tuple
class Animal:
    def __init__(self, name: str, life: int, sound:str=None):
        print("동물을 생성합니다.")
        self.name = name
        self.sound = sound
        self.life = life
    
    def do_nothing(self):
        print("아무것도 하지 않습니다.")
    
    def bark(self):
        print(f"{self.name}는 {self.sound} 소리를 냅니다.")
    
    def info(self):
        if self.life == None:
            print(f"{self.name}의 기대수명은 아직 알려지지 않습니다.")
        else:
            print(f"{self.name}의 기대수명은 {self.life} 살입니다.")
```

```python
dog = Animal(1, 15)
```

```python
dog.bark()
```

```python
dog.info()
```

```python
whale = Animal(name="고래", sound="고래고래")
```

```python
whale.info()
```

```python
dog.do_nothing()
```

```python
cat = Animal("고양이", "야옹")
```

```python
lion = Animal("사자", "어흥")
```

```python
elephant = Animal("코끼리", "뿌우")
```

```python
animals = [dog, cat, lion, elephant]
animals_dict = {animal.name: animal for animal in animals}
```

```python
while True:
    print("\n울음 소리를 듣고 싶은 동물은 입력하세요")
    for animal in animals:
        print(animal.name, end=" ")
    input_animal = input("동물 이름 입력")
    animals_dict[input_animal].bark()
```

- class 아래에 __init__이라는 함수가 정의되어 있습니다. 이는 Espresso 클래스로 객체를 만들 때 무조건 호출되는 생성자라는 함수입니다.
- python에서 클래스 이름을 지을 때는 첫번째 글자를 대문자, 나머지 글자를 소문자로 해줍니다. 이를 Pascal camel case라고 부르는데, 네이밍 컨벤션과 관련해서는 나중에 따로 모아서 다뤄드리겠습니다. 
- 생성자 함수의 첫번째 파라미터는 self 입니다. 이는 객체 자기자신을 의미합니다. 
- 생성자 함수의 두번째 파라미터부터는 객체를 만들때 필요한 변수들입니다.
- 클래스에 대한 유명한 비유로 "붕어빵을 찍어내는 틀"이 있습니다.

## 상속(inheritance)

상속은 객체지향 프로그래밍에서 아주아주 중요한 개념입니다.  
이는 이미 만들어놓은 클래스를 상속받아서 코드를 재사용하여 손쉽게 다양한 클래스를 만들 수 있도록 해줍니다.  

만약 에스프레소 뿐만 아니라 아메리카노를 만들고 싶다고 가정해보겠습니다.  
아메리카노를 표현하기 위해서는 원두와 샷 이외에도 핫인지 아이스인지, 사이즈는 레귤러인지 그란데인지 벤티인지 등의 정보가 필요합니다.  
아메리카노 뿐만 아니라 카페라떼, 카페모카, 캬라멜 마끼야또 등등 커피의 종류는 많습니다.  
이 때마다 새로 클래스를 만들 경우 불편하겠죠?  
그런데 이 모든 커피들이 다 에스프레소를 기반으로 만드는 것들입니다.  

따라서 에스프레소라는 클래스를 상속받으면 쉽게 구현할 수 있습니다.  
이렇게 상속을 받는 클래스는 자식 클래스, 상속 대상은 부모 클래스라고 부릅니다.  

### 상속 기본 사용법
한번 에스프레소 클래스를 상속 받은 아메리카노 클래스를 만들어 보겠습니다.

- class 정의 시에 상속받고자 하는 클래스 명을 () 안에 넣어주어 상속을 구현합니다.  
- 자식 클래스는 기본적으로 부모 클래스의 인스턴스 변수와 메서드를 갖게 됩니다.
- Americano 클래스에는 __str__ 함수를 구현하지 않았지만, 부모 클래스에 구현해놓았으므로 americano 객체를 출력할 경우 원두와 샷 수를 출력할 것입니다.
- 현재는 일부러 bean과 shot을 인스턴스 변수에 설정하지 않았습니다. 그래서 출력시 에러가 날 것입니다.

- 아메리카노 객체를 만든 뒤, print_info 메서드를 호출해보았습니다.
- 아메리카노 클래스에는 print_info 메서드가 정의되지 않았습니다만, 상속받은 espresso 클래스의 print_info 메서드가 호출되었습니다.
- 바로 이것이 상속을 사용하는 이유입니다. 아메리카노 객체를 사용하다보면 어느 원두를 사용한 에스프레소가 몇 샷 들어갔는지 필요한 순간이 올 것입니다. 이 때, 일일이 새로 구현하는 것이 아니라 부모 클래스에 정의된 메서드를 그대로 사용할 수 있습니다.


### 메서드 오버라이딩
- 그런데 이 때 한가지 문제가 있습니다. 해당 커피의 정보를 출력하는 함수인 print_info 를 호출해보면 아메리카노이지만 여전히 에스프레소로 출력됩니다. 이를 아메리카노를 출력하도록 변경해보겠습니다.

- 자식 클래스인 Americano에서 print_info라는 동일한 함수명에 내용만 변경했습니다.
- 이와 같이 자식 클래스에서 부모의 메서드를 덮어 쓰는 걸 "오버라이딩" 이라고 부릅니다.

- 자 이제 바리스타가 아메리카노 주문도 받을 수 있도록 구현해봅시다.
- 주문 결과 받은 커피의 print_info 함수를 호출해서 어떤 커피를 받았는지 출력해봅시다.
