## matplotlib line chart
가장 기본 그래프인 선 그래프를 그릴 수 있는 다양한 방법들을 알아봅니다.

### line chart 그리기 및 스타일링

경기도에서 서울로 전입한 사람 수를 line chart로 그려보면서 matplotlib의 기본 옵션들을 익혀보겠습니다. 먼저 plt.plot() 함수에 series를 전달하면 series의 index를 x축, series의 값들을 y축으로 하는 선 그래프를 그릴 수 있습니다. 그 외 옵션들은 다음과 같은 역할을 수행합니다.

- plt.figure: 도표 크기 지정
- plt.title: 도표 제목 지정
- plt.xticks: 도표 x축 각 틱별 라벨 지정
- plt.xlabel, plt.ylabel: x축 y축을 각각 설명하는 라벨 지정
- plt.legend: 도표 오른쪽 위에 선그래프를 설명하는 라벨 추가. 이 때, plot 함수 호출시에 라벨을 지정해주어야 함.

```python
from matplotlib import pyplot as plt

plt.figure(figsize=(12, 6))
plt.title("서울 → 경기도  인구 이동")
plt.xticks(size=8, rotation=90, ha='right')
plt.xlabel("기간")
plt.ylabel("이동 인구수")
plt.plot(df_seoul.loc["경기도"], label="서울 → 경기")
plt.legend()
```

### matplotlib 스타일링

matplotlib에서 지원하는 기능들로 그래프를 더 이쁘게 그릴 수 있습니다. 보기 좋은 떡이 먹기도 좋다고 눈에 잘 들어오는 그래프가 더 좋겠죠?

- plt.style.use: 미리 정의된 테마 사용, [테마 링크](https://matplotlib.org/stable/gallery/style_sheets/style_sheets_reference.html)

```python
from matplotlib import pyplot as plt

print(plt.style.available)
```

```python
plt.style.use("ggplot")
plt.figure(figsize=(12, 5))
plt.xticks(size=8, rotation=90, ha='right')
plt.title("서울 → 경기도  인구 이동")
plt.xlabel("기간")
plt.ylabel("이동 인구수")
plt.plot(df_seoul.loc["경기도"], label="서울 → 경기", marker="o", markersize=5)
plt.legend()
```

### 화면 분할
matplotlib에서 여러개의 도표를 한번에 만들고 싶을 때가 있다. 그럴 때는 figure의 subplot 기능을 사용하여 axe 객체들을 만들고, 각 axe별로 그래프를 그려주면 됩니다.

```python
fig = plt.figure(figsize=(12, 8))
fig.tight_layout(pad=10.0)

ax1 = fig.add_subplot(2, 2, 1)
ax2 = fig.add_subplot(2, 2, 2)
ax3 = fig.add_subplot(2, 2, 3)
ax4 = fig.add_subplot(2, 2, 4)

sr_one = df_seoul.loc["경상남도"]
sr_two = df_seoul.loc["경상북도"]
sr_three = df_seoul.loc["전라남도"]
sr_four = df_seoul.loc["전라북도"]

ax1.plot(sr_one, marker="o", markersize=5, label="서울 → 경남")
ax2.plot(sr_two, marker="o", markersize=5, label="서울 → 경북", color="skyblue", markerfacecolor="blue")
ax3.plot(sr_three, marker="o", markersize=5, label="서울 → 전남", color="olive", markerfacecolor="green")
ax4.plot(sr_four, marker="o", markersize=5, label="서울 → 전북", color="yellow", markerfacecolor="orange")

ax1.set_title("서울 → 경남")
ax1.legend()
ax2.set_title("서울 → 경북")
ax2.legend()
ax3.set_title("서울 → 전남")
ax3.legend()
ax4.set_title("서울 → 전북")
ax4.legend()


ax1.set_xticklabels(sr_one.index, rotation=90, size=5)
ax2.set_xticklabels(sr_two.index, rotation=90, size=5)
ax3.set_xticklabels(sr_three.index, rotation=90, size=5)
ax4.set_xticklabels(sr_four.index, rotation=90, size=5)

plt.subplots_adjust(wspace=0.3, hspace=0.3)
```

### Stacked bar graph
bargraph도 area와 마찬가지로 stack 옵션을 적용하여서 그래프를 그릴 수 있습니다. 주로 구성 비율의 변화 등을 보여줄 때 사용합니다.

```python
df_seoul.index
```

```python
destinations = [
    '부산광역시', '대구광역시', '인천광역시', '광주광역시', '대전광역시', '울산광역시', '세종특별자치시',
    '경기도', '강원도', '충청북도', '충청남도', '전라북도', '전라남도', '경상북도', '경상남도', '제주특별자치도'
]
```

```python
target_years = ["2010", "2011", "2012", "2013", "2014", "2015", "2016", "2017"]
```

```python
transposed_df = df_seoul.loc[destinations, target_years].transpose()
```

```python
total = transposed_df.sum(axis=1)
```

```python
total
```

```python
ratio_df = transposed_df.div(total, axis=0)
```

```python
ratio_df
```

```python
first_row = ratio_df.iloc[0]
```

```python
ratio_df = ratio_df.sort_index(key=lambda x: first_row[x], axis=1, ascending=False)
```

```python
import matplotlib.colors as mcolors
import random

def generate_color_palette(num_colors):
    all_colors = list(mcolors.CSS4_COLORS.keys())
    random.shuffle(all_colors)
    color_subset = all_colors[:num_colors]
    return color_subset
```

```python
ratio_df.columns
```

```python
ax = ratio_df.plot(
    kind="bar", 
    stacked=True,
    figsize=(12, 5),
    width=0.5,
    color=generate_color_palette(len(ratio_df.columns))
)
legend = ax.legend(loc='lower left', bbox_to_anchor=(-0.1, -0.25), ncol=8)
plt.title("2010년대 서울 전출 인구 지역 비율")
plt.xticks(size=8, rotation=0)
```
