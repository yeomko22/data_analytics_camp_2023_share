## Practice 3. SARIMA 

데이터 셋 링크: https://www.kaggle.com/datasets/robervalt/sunspots

태양 흑점 데이터 셋은 강한 계절성을 포함하는 데이터 셋입니다. 태양의 흑점은 11년을 주기로 계절성을 갖습니다. 월별로 측정한 데이터는 너무 많기 때문에 연도별 평균 값을 낸 데이터 셋으로 리샘플링 하겠습니다.

```python
from utils import load_sunspots

train_df, test_df = load_sunspots()
```

```python
train_df = train_df["Sunspots"].resample("Y").mean().to_frame()
test_df = test_df["Sunspots"].resample("Y").mean().to_frame()
```

```python
train_df.plot(figsize=(10, 2))
test_df.plot(figsize=(10, 2))
```

계절성을 확인하기 위해서 acf plot을 그려보면 11년을 주기로 가지고 있음을 알 수 있습니다.

```python
from utils import plot_acf_pacf

plot_acf_pacf(train_df, "Sunspots");
```
